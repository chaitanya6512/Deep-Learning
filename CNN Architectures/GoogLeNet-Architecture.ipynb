### What is GoogLeNet?
- GoogLeNet also known as Inception Net, is a convolutional neural network(CNN) developed by researchers at Google.
- It is a 22 layer deep architecture and was trained on the ImageNet dataset.
- It can classify objects into 1,000 different categories.
### Architecture:
- The GoogLeNet architecture is based on building a deeper model to acheive greater accuracy while keeping it computationally efficient,
- Neural networks with deeper architectures can capture complex patterns and extract hierarchical features which helps in generalizing better to new, unseen data.
#### 1. Global average pooling:
- It is used at the end of the GoogLeNet architecture before fully connected layer to reduce the spatial dimensions of input, which helps reduce computational costs.
- It also helps reduce overfitting.
#### 2. Inception module:
- It uses multiple filter sizes (5x5, 3x3, 1x1) to extract features at different scales and concatenates their output into a single output.
#### 3. Auxiliary classifier:
- When training datasets on a deep leaning model, the common problem is the vanishing gradient problem.
- Auxiliary classifiers are used at intermediate layers of the GoogLeNet architecture to prevent the vanishing gradient problem during training.
### Key features:
#### 1. Efficiency:
- The architecture of GoogLeNet consists of many 1x1 filters, which reduce the number of parameters and make the model highly efficient.
#### 2. Accuracy:
- GoogLeNet achieved state-of-the-art accuracy on the ImageNet Large Scale Visual Recognition Challenge(ILSVRC) dataset in 2014.
#### 3. Vanishing gradient:
- GoogLeNet uses a auxiliary classifier to address the vanishing gradient problem.
#### 4. Transfer learning:
- GoogLeNet is trained on ImageNet dataset and can classify 1,000 objects.
- This can be used to boost the performance of a related task.
