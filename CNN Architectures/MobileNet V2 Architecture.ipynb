### MobileNet V2:
- MobileNet V2 is a higly efficent convolutional neural network architecture designed for Mobile and Embedded vision applications.
- MobileNetV2 improves upon predecessor, MobileNetV1, by providing better accuracy and reducing computational complexity.
### Key features of MobileNetV2:
#### 1. Inverted Residuals:
- MobileNetV2 intriduces the inverted resdiuals with linear bottlenecks.
- This approach preserves the input and output dimensions while performing the intermediate layers in a lower-dimensional space, reducing computational cost.
- The inverted residual block consists of three layers -
  1. 1x1 convolution(Expansion layer) - Expands the input channels by a factor, increasing the dimensionality of the data.
  2. Depthwise Convolution - Applies depthwise convolution to each expanded channel independently, performing spatial convolution.
  3. 1x1 convolution(Projection layer) - Projects the expanded data back to a lower-dimensional space, reducing the number of channels to the output size.
#### 2. Depthwise Seperable Convolution:
- Similar to MobileNetV1, MobileNetV2 utilizes the depthwise seperable convolution.
- Which splits a standard convolution into two operations.depthwise convolution and pointwise convolution.
- This seperation significantly reduces the number of paramters and computations, making the network more efficient.
#### 3. Linear Bottlenecks:
- The architecture incorportes the linear bottlenecks between layers, ensuring the manifold of the input data is not overly compressed.
- This technique helps in retaining more information and improved accuracy.
- The linear bottlenecks follows the pattern of 1x1 convolution for expansion, depthwise convolution for spatial filtering and another 1x1 convolution for projection.
#### 4. ReLU6 Activation Function:
- MobileNetV6 employs the ReLU6 activation function, amodified version of the ReLU function.
- ReLU restricts the activation values to a range[0, 6], providing better quantization properties for efficient computation on mobile devices.
- This function helps in achieving a balance between accuracy and efficient.

### Implementing MobileNetV2 using TensorFlow:

```python
# Importing Required Libraries
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
import numpy as np

# Load the MobileNetV2 Model
model = MobileNetV2(weights='imagenet')

# Load an image for testing
img_path = 'Image-1.jpg'
img = image.load_img(img_path, target_size=(224, 224))

# Preprocess the image
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Make Predictions
pred = model.predict(x)
print('Predicted:', decode_predictions(pred, top=3)[0])
```

- The output contain a list of tuples. Each tuple contains three elements.
  1. ClassID - A unique identifier for the predicted class.
  2. Class Name - The human-readable label for the predicted class.
  3. Probability Score - The confidence level of the model for that prediction, expressed as a probability.
