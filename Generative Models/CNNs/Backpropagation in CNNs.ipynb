### What is Backpropagation?
- Backpropagation short for "Backpropagation of errors" is an algorithm used to calculate the gradient of the loss function of a neural network with respect to its weights.
- It is essentially a method to update the weights to minimize the loss.
- Backpropagation is crucial because it tells us how to change our weights to improve our network's performance.

### Fundamentals of Backpropagation:
- The process involves three main steps: the forward pass, loss calculation, and the backward pass.
#### 1. Forward Pass:
- During the forward pass, input data is passed through the network to compute the output.
- For a CNNs, this involves several key operations:
  1. Convolutional Layers: Each convolutional layer applies numerous filters to the input. For agiven layer "l" with filters denoted by "F", input "I", and bias "b", the output "O" is given by: O=(I*F)+b. Here "*" denotes the convolutional operation.
  2. Activation Function: After convolution, an activation function σ \sigma σ(e.g ReLU) is applied element-wise to introduce non-linearity: O=σ((I*F)+b).
  3. Pooling Layers:Reduces dimensionality, summarizing the features extracted by the convolutional layers.
#### 2. Loss Calculation:
- After computing the output, a loss function is calculated to assess the error in prediction.
- Common loss function include mean squared error for regression tasks or cross-entropy for classification.
  - L = -∑ylog(y^).
  - Here y is the true label and y^ is the predicted label.
#### 3. Backward Pass(Backpropagation):
- The backpropagation computes the gradient of loss function with respect to each weight in the network by applying the chain rule:
  1. Gradient with respect to output: First, calculate the gradient of the loss function with respect to the output of the final layer.∂L/∂O.
  2. Gradient through activation function: Apply the chain rule through the activation function: ∂L/∂I = ∂L/∂O.∂O/∂I For ReLU, ∂O/∂I is 1 for I >0 and 0 otherwise.
  3. Gradient with respect to filters in convolutional layers: Continue applying the chain rule to find the gradients with respect to the filters: ∂L/∂F = ∂L/∂O * rot180(I). Here, rot180(I) rotates the input by 180 degrees, aligning it for the convolution operation used to calculate the gradient with respect to the filter.

#### Weight Update:
- Using the gradients calculated, the weights are updated using an optimization algorithm such as SGD:
  - F(new) = F(old) - η ∂L/∂F.
  - where,
    - η is the learning rate, which controls the step size during the weight update.
