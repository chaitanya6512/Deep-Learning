### Bidirectional LSTM:
- BiLSTM is an extension version of traditional LSTM network.
- Unlike convolutional LSTM that process sequences in only one direction.
- BiLSTM allow information to flow from both forward and backward enabling them to capture more contextual information.
- This makes BiLSTMs particularly effective for tasks where understanding both past and future context is crucial.
### Understanding Bidrectional LSTM(BiLSTM):
- A Bidirectional LSTM consists of two separate LSTM layers.
  1. Forward LSTM - Process the sequence from start to end.
  2. Backward LSTM - Process the sequence from end to start.
- The outputs of both LSTMs are then combined to form the final output.
- Mathematically, the final output at time t is computed as - p(t) = p(t(f)) + p(t(b)).
- where,
  - p(t) - Final probability of the network
  - p(t(f)) - Probability vector from the forward LSTM network.
  - p(t(b)) - Probability vector from the backward LSTM network.
